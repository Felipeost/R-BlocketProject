From this part is the actual extraction of data from SCB

```python
# Extracting data the easiest way using JSON query and link from SCB:s website
import requests
import json

session = requests.Session()

query = {
  "query": [
    {
      "code": "Region",
      "selection": {
        "filter": "item",
        "values": [
          "00"
        ]
      }
    },
    {
      "code": "Agarkategori",
      "selection": {
        "filter": "item",
        "values": [
          "000"
        ]
      }
    }
  ],
  "response": {
    "format": "json"
  }
}

url = "https://api.scb.se/OV0104/v1/doris/sv/ssd/START/TK/TK1001/TK1001A/PersBilarA"

response = session.post(url, json=query)
response_json = json.loads(response.content.decode('utf-8-sig'))
car_per_region_data = response_json
```

```python
# Viewing the data
car_per_region_data
```

```python
# Gathering the amount of passenger cars and the years from the data. Then grouping it using pandas to get total cars per year.
# Remvoing the year 2023 because of incomplete data.
import pandas as pd

years = []
cars = []
for entry in car_per_region_data['data']:
       if entry['values'] != ['..']:
            years.append(int(entry['key'][-1]))
            cars.append(int(entry['values'][0]))
           
# Create a DataFrame
df_cars = pd.DataFrame({'Year': years, 'Total Cars': cars})

# Calculate total cars for each year
total_cars_per_year = df_cars.groupby('Year')['Total Cars'].sum().reset_index()

print(total_cars_per_year)
```

```python
#Plotting data
from matplotlib import pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(total_cars_per_year['Year'], total_cars_per_year['Total Cars'], marker='o', color='b', linestyle='-')
plt.ylabel('Total Cars (Millions)')
plt.title('Total amount of cars per year in Sweden')
plt.xticks(total_cars_per_year['Year'], rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
